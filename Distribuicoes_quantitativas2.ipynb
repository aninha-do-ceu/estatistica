```python
plt.plot(x, norm.pdf(x, loc=2, scale=0.5))
plt.show()
```

```python
x = np.linspace(-1,1,1000)
```

```python
fig, ax = plt.subplots(ncols = 2, figsize = (10, 6))
ax[0].set_title('Distribuição Normal: F.D.P')
ax[0].plot(x, [norm.pdf(i, 0, 1) for i in x])
ax[1].set_title('Distribuição Normal: Probabilidade Acumulada')
ax[1].plot(x, [norm.cdf(i, 0, 1) for i in x])
```

```python
x = np.linspace(-1, 1, 1000)

for i in x:
    prob = norm.cdf(i, 0, 1)
    if(abs(2/3 - prob) < 10**-3):
        print(f'{i} = {prob}')
```
